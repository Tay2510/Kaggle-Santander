# Preprocess Data

```python
# Get the target
test_ID = test_df['ID']
y = np.log1p(train_df.target)

train_df.drop(["ID", "target"], axis=1, inplace=True)
test_df.drop(["ID"], axis=1, inplace=True)

train_df.head()
```

```python
# logarithmic
train_df.loc[:, :] = np.log1p(train_df.values)
test_df.loc[:, :] = np.log1p(test_df.values)

train_df.head()
```

```python
# Scale values. Since it's time-series we shouldn't scale column-wise
mean = np.mean(train_df.values)
std = np.std(train_df.values)

train_df.loc[:, :] = ((train_df.values - mean) / std)
test_df.loc[:, :] = ((test_df.values - mean) / std)

train_df.head()
```

# LSTM with Keras

```python
from sklearn.model_selection import train_test_split
from keras.callbacks import *
from keras.layers import *
from keras.optimizers import *
from keras.callbacks import *
from keras.models import Model, Sequential

val_fraction = 0.2

# Split into train & test
X_train, X_val, y_train, y_val = train_test_split(
    train_df[ordered_cols],
    y.values, 
    test_size=val_fraction,
    random_state=42,
    shuffle=True
)

# Loss function
def rmse(y_true, y_pred):
    return K.sqrt(K.mean(K.square(y_pred - y_true), axis=-1)) 
```

```python
from keras.regularizers import L1L2

reg_none = L1L2(l1=0.0, l2=0.0)
reg_L1 = L1L2(l1=0.01, l2=0.0)
reg_L2 = L1L2(l1=0.0, l2=0.01)
reg_L1L2 = L1L2(l1=0.01, l2=0.01)

def build_LSTM_model(reg, unit=64):
    model = Sequential()
    model.add(LSTM(unit, return_sequences=True, input_shape=(len(ordered_cols), 1)))
    model.add(LSTM(unit, return_sequences=True, kernel_regularizer=reg))
    model.add(LSTM(int(unit/2)))
    model.add(Dense(1, activation='linear'))
    return model
```

# Load Models

```python
from keras.models import load_model
model = load_model('model.h5', custom_objects={'rmse': rmse})
model_L1 = load_model('model_L1.h5', custom_objects={'rmse': rmse})
model_L2 = load_model('model_L2.h5', custom_objects={'rmse': rmse})
model_L1L2 = load_model('model_L1L2.h5', custom_objects={'rmse': rmse})
```

```python
model.summary()
```

# Ensemble and Predict

```python
test_df.head()

model_to_ensemble = [model, model_L1, model_L2, model_L1L2]

ensemble = None

for m in model_to_ensemble:
    pred = np.expm1(model.predict(np.expand_dims(test_df[ordered_cols].values, 2)))
    
    if ensemble is None:
        ensemble = pred
    else:
        ensemble = ensemble + pred

ensemble = ensemble / len(model_to_ensemble)

sub = pd.DataFrame()
sub['ID'] = test_ID
sub['target'] = ensemble
sub.to_csv('submission_lstm.csv',index=False)
```
