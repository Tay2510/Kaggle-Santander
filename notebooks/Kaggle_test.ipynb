# Preprocessing

```python
import sklearn
```

```python
df_array = df_train.values
```

```python
df_array.shape
```

```python
df_X = [list(d)[2:] for d in df_array]
df_y = [d[1] for d in df_array]
```

```python
X = np.array(df_X, dtype=np.float32)
y = np.array(df_y, dtype=np.float32)
```

```python
X.shape
```

```python
y.shape
```

```python
X[0]
```

## split for train/val

```python
val_fraction = 0.2
val_split_idx = int(val_fraction * len(y))

X_train = X[val_split_idx:]
y_train = y[val_split_idx:]
X_val = X[:val_split_idx]
y_val = y[:val_split_idx]

assert len(X_train) + len(X_val) == len(X)
assert len(y_train) + len(y_val) == len(y)

print('Train:', len(X_train), 'Val:', len(X_val))
```

# Keras for MLP training

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
model = Sequential()
model.add(Dense(64, input_dim=X.shape[1], activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=20, batch_size=128)
```
